```python
import altair as alt
import pandas as pd

from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
trem_list = pd.read_pickle('../data/tremor/aghosh.pkl')
```

```python
tremors = trem_list[0]
tremors.head()
```

```python
timestamp = pd.to_datetime(tremors[['Year', 'Month', 'Day', 'Hour', 'Minute']])
type(timestamp)
```

```python
ts = timestamp.to_frame(name='Time')
type(ts)
```

```python
cleanup = tremors.drop(['Year', 'Month', 'Day', 'Hour', 'Minute'], axis=1)
cleanup.head()
```

```python
timed_tremors = pd.concat([ts, cleanup], axis=1)
timed_tremors.head()
```

```python
timed_tremors.to_pickle("reducedDF.pkl")
```

```python
test = pd.read_csv('../data/geodata/slab_contours.csv')
```

```python
test.head()
```

```python
test = test[test.Depth_km != -55]
test.Depth_km.unique()
```

```python
test2 = test.loc[test['Depth_km'] == -5]
```

```python
test3 = test2.drop(['Depth_km'], axis=1)
test3.head()
```

```python
type(test3.X)
test3.X.tolist()
```

```python
test3.plot.line(x='X', y='Y')
```

```python
ax=plt.gca()

plt.gca().set_aspect('equal', adjustable='box')
plt.draw()
for i in test.Depth_km.unique():
    test2 = test.loc[test['Depth_km'] == i]
    test3 = test2.drop(['Depth_km'], axis=1)
    test3.plot.line(x='X', y='Y',ax=ax,legend=False,color='k')
```

```python
plt.figure(figsize=(8,16))
ax2 = plt.gca()
test2 = test.loc[test['Depth_km'] == -20]
test3 = test2.drop(['Depth_km'], axis=1)

m = Basemap(width=250000,height=400000,projection='lcc',
            resolution='i',lat_1=47,lat_2=48,lat_0=48,lon_0=-123)
# draw coastlines.
m.drawcoastlines(ax=ax2)

for i in test.Depth_km.unique():
    test2 = test.loc[test['Depth_km'] == i]
    test3 = test2.drop(['Depth_km'], axis=1)
    x, y = m(test3.X.tolist(), test3.Y.tolist())  # transform coordinates
    m.plot(x, y, ax=ax2, color=cm.winter(1+i/90))

# test3.plot.line(x='X', y='Y',ax=ax2,legend=False,color='k')
```
